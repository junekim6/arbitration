```python
import pandas as pd

# Show all columns in dataframe
pd.set_option('display.max_columns', None)

# Show full content of columns
pd.set_option('display.max_colwidth', None)
```

```python
icsid = pd.read_csv('icsid_database.csv')
unctad = pd.read_excel('UNCTAD_modified.xlsx')
```

```python
display('ICSID data')
display(icsid.shape)
display(icsid.sample(3))

display('UNCTAD data')
display(unctad.shape)
display(unctad.sample(3))
```

```python
unctad[unctad['FULL CASE NAME'].str.contains('Coöperatief')].head(1)
```

```python
unctad['case_number'] = unctad['FULL CASE NAME'].str.extract(r'.*Case No\.(.*)\)', expand=False)
```

```python
unctad = unctad[unctad.case_number.notnull()]
```

```python
icsid["case_number"] = icsid["case_number"].astype(str).str.strip()
unctad["case_number"] = unctad["case_number"].astype(str).str.strip()
```

```python
# join unctad and icsid data based on "case_number" column

joined = icsid.merge(unctad, on='case_number', how='outer')
```

```python
unmatched_unctad = joined[joined["claimant"].isna()].reset_index(drop=True)
unmatched_unctad.to_csv("unmatched_unctad.csv", index=False)

display(unmatched_unctad.shape)
```

```python
unmatched_icsid = joined[joined["NO."].isna()].reset_index(drop=True)
unmatched_icsid.to_csv("unmatched_icsid.csv", index=False)

display(unmatched_icsid.shape)
```

```python
matched = joined[~(joined["claimant"].isna() | joined["NO."].isna())].reset_index(drop=True)
```
